```python
import requests
import pandas as pd
```

```python
url = 'http://localhost:5000'
```

```python
resp = requests.get(f"{url}/apispec_1.json").json()

endpoints = list(resp["paths"].keys())
print(endpoints)
```

```python
categories_stats = f"{url}/api/v1/stats/categories"

resp = requests.get(categories_stats).json()

df_cat_stats = pd.DataFrame()

for k, v in resp.items():
    tmp = pd.json_normalize(v)
    tmp['category'] = k

    df_cat_stats = pd.concat([df_cat_stats, tmp])

df_cat_stats
```
